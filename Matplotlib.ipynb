```python
#Pie Chart >> To Understand the Percentahge Distribution of Data
```

```python
langs = ["C","Python","Java"]
students=[20,100,40]
plt.pie(students,labels=langs,autopct="%1.1f%%")
plt.show()
```

```python
explode=(0.0,0.1,0.1)
langs = ["C","Python","Java"]
students=[20,100,40]
plt.pie(students,labels=langs,autopct="%1.1f%%",explode=explode)
plt.show()
```

```python
explode=(0.0,0.1,0.1)
langs = ["C","Python","Java"]
students=[20,100,40]
plt.pie(students,labels=langs,autopct="%1.1f%%",explode=explode,shadow=True)
plt.show()
```
